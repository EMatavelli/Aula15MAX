# Fluxo Básico de Análise

Exemplo completo de análise utilizando Pandas.

```
import pandas as pd
import matplotlib.pyplot as plt

# Ler arquivo
df = pd.read_csv('arquivo.csv')

# Explorar dados
df.head()
df.info()

# Estatísticas
df.describe()

# Filtrar
df[df['idade'] > 18]

# Agrupar
df.groupby('cidade')['vendas'].sum()

# Gráfico
df['cidade'].value_counts().plot(kind='bar')
plt.show()
```
